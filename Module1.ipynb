```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
batting = pd.read_csv('Batting2024.csv')
fielding = pd.read_csv('Fielding2024.csv')
pitching = pd.read_csv('pitching2024.csv')
```

```python
def_data = pd.merge(pitching, fielding, on='Tm', how='outer')

teams = batting[['Tm']].copy()

team_offense = batting[['Tm', 'R/G', 'OBP', 'SLG', 'BA']].copy()
team_defense = def_data[['Tm', 'ERA', 'WHIP','RA/G_x', 'Fld%']].copy()
team_defense = team_defense.rename(columns={'RA/G_x': 'RA/G'})

team_data = pd.merge(team_offense, team_defense, on='Tm', how='outer')
print(team_data)
```

```python
display(team_data)
```

```python
fig, ax = plt.subplots()
ax.scatter(team_data['R/G'], team_data['BA'])


ax.set_title('Team Relationship between Runs per Game and Batting Average')
ax.set_xlabel('Runs per Game')
ax.set_ylabel('Batting Average')

plt.show()
```
