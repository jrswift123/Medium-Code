```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
steam_data = pd.read_csv('steam_chart.csv')
```

```python
steam_data.head(10)
```

```python
def find_similar_row_names(df, target_index, name_column='name', top_n=10):
    string_cols = [col for col in df.select_dtypes(include=['object']).columns 
                  if col != name_column]
    df['combined_text'] = df[string_cols].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)
    
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(df['combined_text'])
    
    cosine_sim = cosine_similarity(tfidf_matrix)
    
    similar_indices = cosine_sim[target_index].argsort()[-top_n-1:-1][::-1]
    
    return df.iloc[similar_indices][name_column].tolist()
```

```python
game_index = 0

similar_names = find_similar_row_names(steam_data, target_index=game_index)

print(steam_data.loc[game_index, 'name'])
for i, name in enumerate(similar_names, 1):
    print(f"{i}. {name}")
```
